```python
# Criando um modelo automatico que analise os clientes com base
# NOTA DE CRÉDITO DELE = SCORE DE CRÉDITO
# Se o cliente é BOM,     OK    ou RUIM
# Se o cliente é Good, Standard ou Poor

# pip install pandas scikit-learn
```

```python
# Vizualizando a base de dados de Clientes
import pandas as pd

tabela = pd.read_csv("clientes.csv")

display(tabela)
display(tabela.info())
```

```python
# Vamos preparar a base de dados para a Inteligência Artificial
# Primeiro precisamos transformar os dados que estão em texto em numeros,
# pois os modelos de IA trabalham melhor com valores numéricos.

# Para isso usamos o LabelEncoder do scikit-learn.
# Ele converte categorias de texto em numeros.

from sklearn.preprocessing import LabelEncoder

codificador_profissao = LabelEncoder()
tabela["profissao"] = codificador_profissao.fit_transform(tabela["profissao"])

codificador_mix = LabelEncoder()
tabela["mix_credito"] = codificador_mix.fit_transform(tabela["mix_credito"])

codificador_comportamento = LabelEncoder()
tabela["comportamento_pagamento"] = codificador_comportamento.fit_transform(tabela["comportamento_pagamento"])

display(tabela)
```

```python
# Agora vamos separar os dados que serão usados para treinar a IA
# e os dados que serão usados para testar o modelo.

# y = variável que queremos prever (score de credito)
# x = variáveis que serão usadas para fazer a previsão

from sklearn.model_selection import train_test_split

y = tabela["score_credito"]

x = tabela.drop(columns=["score_credito", "id_cliente"])

# Aqui dividimos os dados em treino e teste
# 70% para treino e 30% para teste

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# Agora vamos treinar os modelos de Inteligência Artificial
# O objetivo é prever o score de credito: BOM, OK ou RUIM

# Importando dois modelos diferentes para testar
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criando os modelos
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# Treinando os modelos com os dados de treino
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Agora vamos testar os modelos para ver qual tem melhor desempenho

previsao_arvore = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# Calculando a taxa de acerto de cada modelo
from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste, previsao_arvore))
display(accuracy_score(y_teste, previsao_knn))
```

```python
# O modelo que teve melhor resultado foi o de Árvore de Decisão
# Agora vamos usar ele para prever o score de credito de novos clientes

novos_clientes = pd.read_csv("novos_clientes.csv")

# Transformando os dados de texto em números novamente
novos_clientes["profissao"] = codificador_profissao.fit_transform(novos_clientes["profissao"])
novos_clientes["mix_credito"] = codificador_mix.fit_transform(novos_clientes["mix_credito"])
novos_clientes["comportamento_pagamento"] = codificador_comportamento.fit_transform(novos_clientes["comportamento_pagamento"])

# Fazendo a previsão
previsao = modelo_arvoredecisao.predict(novos_clientes)

display(novos_clientes)
display(previsao)
```
